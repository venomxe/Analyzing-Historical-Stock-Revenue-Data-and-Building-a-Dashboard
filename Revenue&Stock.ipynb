```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")
```

```python
tesla_revenue_list = []
tables = soup.find_all("table")
```

```python
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        rows = table.find_all("tr")
        for row in rows[1:]:
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip()
                if revenue:
                    tesla_revenue_list.append({"Date": date, "Revenue": revenue})
        break
```

```python
if tesla_revenue_list:
    tesla_revenue = pd.DataFrame(tesla_revenue_list)
    tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace("$", "", regex=False).str.replace(",", "", regex=False)
    tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"])
```

```python
tesla_revenue.tail()
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                        vertical_spacing=0.1,
                        subplot_titles=(f"{stock} Stock Price", f"{stock} Revenue"),
                        row_width=[0.4, 0.6])

    fig.add_trace(go.Scatter(
        x=stock_data['Date'],
        y=stock_data['Close'],
        name='Stock Price',
        mode='lines'
    ), row=1, col=1)

    fig.add_trace(go.Scatter(
        x=revenue_data['Date'],
        y=revenue_data['Revenue'],
        name='Revenue',
        mode='lines'
    ), row=2, col=1)

    fig.update_layout(height=600, width=1000, title_text=f"{stock} Stock Price and Revenue", showlegend=False)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Stock Price (USD)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue (USD)", row=2, col=1)
    fig.show()
```

```python
make_graph(tesla_data, tesla_revenue, 'Tesla')
```

```python
make_graph(gme_data, gme_revenue, 'GameStop')
```
